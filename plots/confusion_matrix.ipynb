```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import ast

```

```python
# Read the data once
df = pd.read_csv("CML_prob.csv")

# Create a list of models
models = ['Logistic Regression', 'SVM', 'KNN', 'Decision Tree', 'Random Forest', 'XGboost', 'Neural Network']

# Determine the number of rows and columns for subplots
num_rows = 4
num_cols = 3

# Create subplots with a gray background
fig, axes = plt.subplots(num_rows, num_cols, figsize=(5 * num_cols, 4 * num_rows))

# Flatten the 2D array of axes for easy iteration
axes = axes.flatten()

# Iterate over models and plot confusion matrix
for i, model in enumerate(models):
    y_true = ast.literal_eval(df.loc[i, 'y_true'])
    y_pred = ast.literal_eval(df.loc[i, 'y_predicted'])

    # Create confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Create a heatmap using seaborn
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'],
                ax=axes[i], annot_kws={'fontsize': 15})  # Adjust the fontsize as needed
    
    axes[i].set_title(f'{model} ', fontsize=15)
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')


# Remove empty subplots
for i in range(len(models) + 3, num_rows * num_cols + 1):
    fig.delaxes(axes[i - 1])
fig.delaxes(axes[8])
fig.delaxes(axes[7])
# Add a note under the subplots
fig.text(0.5, 0.25, 'Internal Validation test', ha='center', fontsize=18, va='center')

# Adjust layout
plt.subplots_adjust(hspace=0.5)

plt.show()
```

```python
multi = pd.read_csv('zero_shot_by_hand - Sheet1.csv')
multi = multi.replace('survive', 1)
multi = multi.replace('die', 0)
y_true = multi[['Inhospital Mortalit(TRUE)']]
mixtral = multi[['Mixtral-8x7B-Instruct-v0.1 ']]
llama370= multi[['Llama-3-70B']]
mistral7b=multi[['Mistral-7B-Instruct']]
llama38=multi[['-Llama-3-8B']]
gpt_4o = multi[['gpt-4o-2024-05-13_outcome']]
gpt_35_turbo = multi[['gpt-3.5-turbo-0125_outcome']]
gpt_4_turbo = multi[['gpt-4-turbo-2024-04-09_outcome']]
gpt_4 = multi[['gpt-4-0613_outcome']]
```

```python

# Determine the number of rows and columns for subplots
num_rows = 3
num_cols = 3

# Create subplots with a gray background
fig, axes = plt.subplots(num_rows, num_cols, figsize=(5 * num_cols, 4 * num_rows))

# Flatten the 2D array of axes for easy iteration
axes = axes.flatten()
# Create confusion matrix for zero-shot classification
cm_zero_shot = confusion_matrix(y_true, mixtral)
plt.subplot(num_rows, num_cols, 1 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('Mixtral-8x7B-Instruct-v0.1')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, llama370)
plt.subplot(num_rows, num_cols, 2 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('Llama-3-70B')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, mistral7b)
plt.subplot(num_rows, num_cols, 3 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('Mistral-7B-Instruct')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, llama38)
plt.subplot(num_rows, num_cols, 4 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('-Llama-3-8B')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, gpt_4o)
plt.subplot(num_rows, num_cols, 5 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('gpt-4o')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, gpt_35_turbo)
plt.subplot(num_rows, num_cols, 6 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('gpt-3.5-turbo')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, gpt_4_turbo)
plt.subplot(num_rows, num_cols, 7 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('gpt-4-turbo')
plt.xlabel('Predicted')
plt.ylabel('Actual')

cm_zero_shot = confusion_matrix(y_true, gpt_4)
plt.subplot(num_rows, num_cols, 8 )
sns.heatmap(cm_zero_shot, annot=True, fmt='d', cmap='Reds', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'])
plt.title('gpt-4')
plt.xlabel('Predicted')
plt.ylabel('Actual')
# Add a note under the subplots
fig.text(0.5, 0.02, 'Zero shot classification', ha='center', fontsize=18, va='center')

fig.delaxes(axes[8])
# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
