```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import ast

```

```python
# Internal validation 

import matplotlib.pyplot as plt

# Set the font sizes
SMALL_SIZE = 30
MEDIUM_SIZE = 32
BIGGER_SIZE = 40

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

# Rest of your code...

# Read the data once
df = pd.read_csv("CML_prob.csv")

# Create a list of models
models = ['a) Logistic Regression', 'b) SVM', 'c) KNN', 'd) Decision Tree', 'e) Random Forest', 'f) XGboost', 'g) Neural Network']

# Determine the number of rows and columns for subplots
num_rows = 4
num_cols = 3

# Create subplots with a gray background
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10 * num_cols, 8 * num_rows))

# Flatten the 2D array of axes for easy iteration
axes = axes.flatten()

# Iterate over models and plot confusion matrix
for i, model in enumerate(models):
    y_true = ast.literal_eval(df.loc[i, 'y_true'])
    y_pred = ast.literal_eval(df.loc[i, 'y_predicted'])

    # Create confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Create a heatmap using seaborn
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Died', 'Survived'], yticklabels=['Died', 'Survived'],
                ax=axes[i], annot_kws={'fontsize': 15})  # Adjust the fontsize as needed
    
    axes[i].set_title(f'{model} ', fontsize=BIGGER_SIZE)
    axes[i].set_xlabel('Predicted', fontsize=SMALL_SIZE)
    axes[i].set_ylabel('Actual', fontsize=SMALL_SIZE)

# Remove empty subplots
for i in range(len(models) + 3, num_rows * num_cols + 1):
    fig.delaxes(axes[i - 1])
fig.delaxes(axes[8])
fig.delaxes(axes[7])
# Add a note under the subplots
#fig.text(0.5, 0.25, 'Internal Validation test', ha='center', fontsize=BIGGER_SIZE, va='center')
plt.subplots_adjust(hspace=0.7, wspace=0.7)
# Adjust layout
#plt.subplots_adjust(hspace=0.5)

plt.show()
```
